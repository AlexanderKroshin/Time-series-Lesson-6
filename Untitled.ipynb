```python
#Подключаем библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
import re
import seaborn as sns
import warnings
```

```python
#Читаем текст, оставим только русские буквы и символы пробела.
with open('C:/Users/Александр/4._Случайные_марковские_процессы/tom_1.txt', 'r', encoding="utf-8") as f:
    text = f.read().lower()
text = re.sub('[^а-я]','', text)
```

```python
RUSSIAN = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
POS = {l: i for i, l in enumerate(RUSSIAN)}
```

```python
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))
```

```python
for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
```

```python
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))
```

```python
#Строим матрицу переходных вероятностей
plt.figure(figsize=(14,8))
sns.heatmap(df)
plt.show()
```

```python
#Считаем вероятность встретить строку из четырёх символов, которая начинается на букву «П» и заканчивается на букву «Р».
start = POS['п']
end = POS['р']
prior = 1. / len(RUSSIAN)
proba_sum = 0.
for c1 in RUSSIAN:
    for c2 in RUSSIAN:
        proba = prior * probabilities[start, POS[c1]] * probabilities[POS[c1], POS[c2]] * probabilities[POS[c2], end]
        proba_sum += proba
```

```python
#Используя матрицу переходных вероятностей, сгенерируем слово, начинающееся с буквы «Б» длиной восемь символов.
s = []
start = 'б'
s.append(start)
length = 8
for i in range(length):
    index = POS[s[i]]
    next_char = np.random.choice(list(RUSSIAN), p=probabilities[index])
    s.append(next_char)

print (''.join(s))
```

```python
#Выберим ряд из предыдущих домашних заданий по временным рядам и оцените его через HMM
path = "C:/Users/Александр/1._Лекция_tsml/Series"
series = pd.read_csv(f"{path}/monthly-sales-of-company-x-jan-6.csv", index_col ='Month', parse_dates = True)["Count"]
```

```python
values = series.values.reshape(-1,1)
```

```python
warnings.filterwarnings('ignore')
```

```python
for n in range(1, 14):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(values)
    print (n, model.score(values))
```

```python
model = GaussianHMM(n_components=10,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
```
